```python
# your code here 1
pull_id_q = """
select *,
CASE
    WHEN INCIDENT_CITY = 'Chicago'
    THEN 'YES'
    ELSE 'NO'
END AS in_chicago
from caseinit
"""

id_d = pd.read_sql_query(pull_id_q, cnx)
id_d.head()
```
